```python
from pathlib import Path
import pandas as pd
import datetime
csv_path = "Crime_Data_2010_2017.csv"
crime_df = pd.read_csv(csv_path)
crime_df.head()
```

```python
clean_crimes = crime_df[["Date Occurred", "Time Occurred", "Area Name", "Reporting District", "Crime Code Description", "Victim Age", "Victim Sex", "Victim Descent", "Premise Description", "Location "]]
clean_crimes.rename(columns= {"Location ": "Location"})
clean_crimes.head()
```

```python
clean_crimes = clean_crimes.dropna(how = "any")
clean_crimes.info()
```

```python
clean_crimes[["Month", "Day", "Year"]] = clean_crimes["Date Occurred"].str.split("/", expand = True)
clean_crimes.head()
```

```python
clean_crimes["Year"] = clean_crimes["Year"].astype("int")
clean_crimes["Month"] = clean_crimes["Month"].astype("int")
clean_crimes["Day"] = clean_crimes["Day"].astype("int")
crime_subset = clean_crimes[clean_crimes["Year"] >= 2015]
crime_subset.info()
```

```python
crime_subset.to_csv("Cleaned_Crime_Data.csv")
```
